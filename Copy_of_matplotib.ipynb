## pie charts


```python
import matplotlib.pyplot as csk
import numpy as ae
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show() 
```

```python
import matplotlib.pyplot as ap
import numpy as op
y = op.array([30, 25, 20, 25])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
ap.pie(y, labels = mylabels)
ap.show() 
```

```python
import matplotlib.pyplot as ad
import numpy as ap
y = ap.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
ad.pie(y, labels = mylabels, explode = myexplode)
ad.show()
```

```python
import matplotlib.pyplot as un
import numpy as ap
y = ap.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["red", "blue", "green", "brown"]
un.pie(y, labels = mylabels, colors = mycolors)
un.show() 
```
